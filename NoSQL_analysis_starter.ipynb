### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and a RatingValue greater than or equal to 4.
query2 = {'LocalAuthorityName': {'$regex': 'London'}, 
          'RatingValue': {'$gte': 4}
         }



# Use count_documents to display the number of documents in the result
count2 = establishments.count_documents(query2)
print(f"The number of documents with London as the Local Authority and a RatingValue greater than or equal to 4 is {count2}.")

# Display the first document in the results using pprint
results=establishments.find(query2)
for doc in results: 
    print(doc)

```

```python
# Convert the result to a Pandas DataFrame
df3=pd.DataFrame(establishments.find(query2))

# Display the first 10 rows of the DataFrame
df3.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.49
longitude = 0.08

query3 = {
    "$and": [
        {"RatingValue": 5, 'geocode.latitude': {'$lte': latitude + degree_search, '$gte': latitude - degree_search}},
        {"RatingValue": 5, 'geocode.longitude': {'$lte': longitude + degree_search, '$gte': longitude - degree_search}}
    ]
}
sort =    [('scores.Hygiene', 1)]


# Print the results
for result in establishments.find(query3).sort(sort).limit(5):
    pprint(result)
```

```python

# Convert result to Pandas DataFrame
df2=pd.DataFrame(establishments.find(query3).sort(sort).limit(5))
df2
```
